```python
dark_horse = cv2.imread('../data/horse.jpg')
show_horse = cv2.cvtColor(dark_horse, cv2.COLOR_BGR2RGB)

rainbow = cv2.imread('../data/rainbow.jpg')
show_rainbow = cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)

bricks = cv2.imread('../data/bricks.jpg')
show_bricks = cv2.cvtColor(bricks, cv2.COLOR_BGR2RGB)
```

```python
def show_histogram(img, title, xmax, ymax=None):
    color = ('b', 'g', 'r')
    for i, clr in enumerate(color):
        hister=cv2.calcHist([img], channels=[i],
                                            mask=None, histSize=[256], ranges=[0, 256])
        plt.plot(hister, color = clr)
        plt.xlim([0, xmax])
        if ymax != None:
            plt.ylim([0, ymax])
    plt.title(title)
```

```python
show_histogram(bricks, 'HISTOGRAM FOR BLUE BRICKS', 256)
```

```python
show_histogram(dark_horse, 'HISTOGRAM FOR DARK HORSE', 256)
```

```python
show_histogram(dark_horse, 'HISTOGRAM FOR DARK HORSE', 15, 50000)
```
