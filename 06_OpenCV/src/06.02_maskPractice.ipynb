```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.segmentation import clear_border
import skimage.morphology as mp
import scipy.ndimage.morphology as sm
```

```python
Airplane = cv2.imread('../data/airplane.bmp', cv2.IMREAD_COLOR)
Airplane = cv2.cvtColor(Airplane, cv2.COLOR_BGR2RGB)
plt.imshow(Airplane);
```

```python
mask = cv2.imread('../data/mask_plane.bmp', cv2.IMREAD_GRAYSCALE)
plt.imshow(mask, cmap='gray');
```

```python
dst = cv2.imread('../data/field.bmp', cv2.IMREAD_COLOR)
dst = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)
plt.imshow(dst);
```

```python
plt.imshow(cv2.copyTo(Airplane, mask, dst));
```
