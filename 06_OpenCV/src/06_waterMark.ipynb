```python
import cv2
import matplotlib.pyplot as plt
```

```python
candyBGR = cv2.imread('../data/candies.png')
candyRGB = cv2.cvtColor(candyBGR, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(candyBGR);
```

```python
plt.imshow(candyRGB);
```

```python
doNotCopy = cv2.imread('../data/watermark_no_copy.png')
doNotCopy = cv2.cvtColor(doNotCopy, cv2.COLOR_BGR2RGB)
plt.imshow(doNotCopy);
```

```python
candyRGB.shape, doNotCopy.shape
```

### Error causes by different image sizes between two images

```python
blended = cv2.addWeighted(src1=candyRGB, alpha=0.5, src2=doNotCopy, beta=0.5, gamma=0)
plt.imshow(blended);
```

```python
doNotCopy = cv2.resize(doNotCopy, (candyRGB.shape[1], candyRGB.shape[0]))
candyRGB.shape, doNotCopy.shape
```

```python
blended = cv2.addWeighted(src1=candyRGB, alpha=0.7, src2=doNotCopy, beta=0.3, gamma=1)
plt.imshow(blended);
```

```python
doNotCopy = cv2.resize(doNotCopy, (100,100))
plt.imshow(doNotCopy);
```

```python
copyLightIMG = candyRGB
waterMark = doNotCopy

x_offset = 0
y_offset = 0
```

```python
x_end = x_offset + waterMark.shape[1]
y_end = y_offset + waterMark.shape[0]

copyLightIMG[y_offset:y_end, x_offset:x_end]=waterMark
```
