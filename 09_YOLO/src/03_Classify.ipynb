result

```python
import matplotlib.pyplot as plt
from PIL import Image
# object detection  rock_scissor_paper
origin_img = Image.open('../data/rock.jpg')
# object detection  
detect_img = Image.open('./runs/classify/predict6/rock.jpg')
# matplotlib  2   
fig, ax = plt.subplots(ncols=2, figsize=(20, 15))
ax[0].imshow(origin_img)
ax[0].set_title('Origin Image')
ax[1].imshow(detect_img)
ax[1].set_title('Detect Image')
plt.show()
```
