```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prophet import Prophet
import seaborn as sns
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
time = np.linspace(0, 1, 365*2)
result = np.sin(2*np.pi*12*time)
ds = pd.date_range('2017-01-01', periods=365*2, freq='D')
df = pd.DataFrame({'ds':ds, 'y':result})
df.head()
```

```python
df['y'].plot(figsize=(10,6));
```

```python
m = Prophet(yearly_seasonality=True, daily_seasonality=True)
```

```python
m.fit(df)
future = m.make_future_dataframe(periods=30)
forecast = m.predict(future)
```

```python
m.plot(forecast)
```

```python
pinkwink_web = pd.read_excel('../data/pinkwink_web.xlsx',
                             names=['ds','hit'])
pinkwink_web.head()
```

```python
df = pd.DataFrame({'ds':pd.to_datetime(pinkwink_web['ds'],
                                        format="%y. %m. %d."),
                    'y':pinkwink_web['hit']})
m = Prophet(yearly_seasonality=True, daily_seasonality=True)
m.fit(df)
future = m.make_future_dataframe(periods=31)
forecast = m.predict(future)
m.plot(forecast);
```

```python
sns.set_style("whitegrid")
sns.set_palette('husl')

plt.figure(figsize=(10,6))
for each in ['trend', 'yhat', 'yhat_lower', 'yhat_upper']:
    plt.plot(forecast['ds'], forecast[each], label=each)

plt.legend(loc='best')
plt.show()
```

```python
plt.figure(figsize=(10,6))
for each in ['trend', 'yhat', 'yearly']:
    plt.plot(forecast['ds'], forecast[each], label=each)

plt.legend(loc='best')
plt.show()
```

```python
m.plot_components(forecast);
```

```python
KIA = yf.download('000270.KS',
                  start='1990-01-01',
                  end='2023-12-31',
                  )
KIA.head()
```

```python
KIA = KIA[["Close"]]
KIA = KIA.reset_index()
KIA.columns = ['ds', 'y']

KIA.head()
```

```python
m = Prophet()
m.fit(KIA)

future = m.make_future_dataframe(periods=300)
future.tail()
```

```python
forecast = m.predict(future)
forecast.head()
```

```python
plt.figure(figsize=(12,6))
plt.plot(KIA['ds'], KIA['y'])
plt.show()
```
