```python
plt.figure()
raw_data['자살생각율'].plot.pie(autopct='%.1f')
plt.ylabel('')
plt.title('자살생각율')
plt.show()
```

```python
f, ax = plt.subplots(1,3, figsize=(14,6))

raw_data['스트레스'].plot.pie(explode=[0,0.01], ax=ax[0], autopct='%.1f')
ax[0].set_title('스트레스를 받은적 있다')

raw_data['우울감경험률'].plot.pie(explode=[0,0.01], ax=ax[1], autopct='%.1f')
ax[1].set_title('우울증을 경험한적 있다')

raw_data['자살생각율'].plot.pie(explode=[0,0.01], ax=ax[2], autopct='%.1f')
ax[2].set_title('자살을 고민한적 있다')

plt.show()
```
