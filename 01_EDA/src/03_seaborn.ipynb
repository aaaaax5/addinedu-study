```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
import seaborn as sns
```

```python
x = np.linspace(0, 14, 100)
y1 = np.sin(x)
y2 = 2*np.sin(x+0.5)
y3 = 3*np.sin(x+1.0)
y4 = 4*np.sin(x+1.5)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

```python
sns.set_style("white")
plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
sns.despine(); plt.show()
```

```python
sns.set_style("dark")
plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

```python
sns.set_style("white")
plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
sns.despine(offset=10)
plt.show()
```

```python
tips = sns.load_dataset("tips")
tips.head(5)
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x=tips["total_bill"])
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips, palette="deep")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="deep")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.swarmplot(x="day", y="total_bill", data=tips, color=".5", palette="deep")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips, palette="deep")
sns.swarmplot(x="day", y="total_bill", data=tips, color=".25", palette="dark")
plt.show()
```

## lmplot, () 

```python
sns.set_style("darkgrid")
sns.lmplot(x="total_bill", y="tip", data=tips, height=7)
plt.show()
```

## lmplot, ()  - hue( )

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, height=7)
plt.show()
```

```python
flights = sns.load_dataset("flights")
flights.head(5)
```

```python
flights = flights.pivot(index="month", columns="year", values="passengers")
flights.head(5)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights, annot=True, fmt="d")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights, annot=True, fmt="d", cmap="YlOrRd")
plt.show()
```

```python
sns.set(style="ticks")
iris = sns.load_dataset("iris")
iris.head(10)
```

```python
sns.pairplot(iris, hue="species")
plt.show()
```

```python
sns.pairplot(iris, x_vars=["sepal_width", "sepal_length"],
             y_vars=["petal_width", "petal_length"],
             hue="species",
             height=3)
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
sns.catplot(x="sex", y="survived", hue="pclass", data=titanic, kind="bar",
            palette="muted", legend=True, height=7)
plt.show()
```
